```python
import matplotlib.pyplot as plt
import numpy as np

# Dane
k_values = [1, 3, 5, 7]
knn_train_acc = [1.000, 0.835, 0.645, 0.584]
knn_test_acc = [0.256, 0.278, 0.299, 0.309]

metrics = ['euclidean', 'manhattan', 'chebyshev', 'cosine']
nc_train_acc = [0.232, 0.217, 0.138, 0.234]
nc_test_acc = [0.221, 0.203, 0.135, 0.221]

gnb_var_smooth = [1e-9, 1e-8, 1e-7, 1e-6]
gnb_train_acc = [0.369, 0.369, 0.369, 0.369]
gnb_test_acc = [0.359, 0.359, 0.359, 0.359]

pca_n_components = [5, 10, 15, 17]
pca_train_acc = [0.288, 0.323, 0.361, 0.362]
pca_test_acc = [0.286, 0.313, 0.352, 0.350]

reg_vals = [0.00, 0.01, 0.10, 1.00]
softmax_train_acc = [0.207, 0.200, 0.173, 0.198]
softmax_test_acc = [0.207, 0.200, 0.172, 0.198]

n_estimators = [10, 50, 100, 200]
rf_train_acc = [0.538, 0.563, 0.569, 0.574]
rf_test_acc = [0.375, 0.394, 0.391, 0.392]

# Tworzenie siatki wykresw
plt.figure(figsize=(16, 12))

# k-NN
plt.subplot(3, 2, 1)
plt.plot(k_values, knn_train_acc, marker='o', label='Train Accuracy')
plt.plot(k_values, knn_test_acc, marker='s', label='Test Accuracy')
plt.title('k-NN: Accuracy vs. k')
plt.xlabel('Liczba ssiadw (k)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()

# Nearest Centroid
plt.subplot(3, 2, 2)
x = np.arange(len(metrics))
plt.plot(x, nc_train_acc, marker='o', label='Train Accuracy')
plt.plot(x, nc_test_acc, marker='s', label='Test Accuracy')
plt.xticks(x, metrics)
plt.title('Nearest Centroid: Accuracy vs. Metric')
plt.xlabel('Metryka odlegoci')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()

# Gaussian NB
plt.subplot(3, 2, 3)
plt.plot(gnb_var_smooth, gnb_train_acc, marker='o', label='Train Accuracy')
plt.plot(gnb_var_smooth, gnb_test_acc, marker='s', label='Test Accuracy')
plt.xscale('log')
plt.title('Gaussian NB: Accuracy vs. var_smoothing')
plt.xlabel('var_smoothing (log scale)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()

# GNB + PCA
plt.subplot(3, 2, 4)
plt.plot(pca_n_components, pca_train_acc, marker='o', label='Train Accuracy')
plt.plot(pca_n_components, pca_test_acc, marker='s', label='Test Accuracy')
plt.title('GNB + PCA: Accuracy vs. PCA components')
plt.xlabel('Number of Components')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()

# Softmax Regression
plt.subplot(3, 2, 5)
plt.plot(reg_vals, softmax_train_acc, marker='o', label='Train Accuracy')
plt.plot(reg_vals, softmax_test_acc, marker='s', label='Test Accuracy')
plt.title('Softmax Regression: Accuracy vs. L2 Regularization')
plt.xlabel('L2 Regularization Strength (reg)')
plt.ylabel('Accuracy')
plt.xticks(reg_vals)
plt.grid(True)
plt.legend()

# Random Forest
plt.subplot(3, 2, 6)
plt.plot(n_estimators, rf_train_acc, marker='o', label='Train Accuracy')
plt.plot(n_estimators, rf_test_acc, marker='s', label='Test Accuracy')
plt.title('Random Forest: Accuracy vs. Number of Trees')
plt.xlabel('Number of Trees')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
