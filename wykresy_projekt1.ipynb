```python
import pandas as pd
df=pd.read_csv("train_data.csv",sep=";")

df['Stay_cleaned'] = df['Stay'].replace({
    '41-50': '40+',
    '51-60': '40+',
    '61-70': '40+',
    '71-80': '40+',
    '81-90': '40+',
    '91-100': '40+',
    'More than 100 Days': '40+'
})
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Admission_Deposit', fill=True, color='skyblue')

plt.title("Krzywa gęstości wpłat przy przyjęciu")
plt.xlabel("Wpłata przy przyjęciu")
plt.ylabel("Gęstość")
plt.tight_layout()
plt.show()

order = df['Stay_cleaned'].value_counts().index.tolist()
plt.figure(figsize=(10, 6))
sns.countplot(x='Stay_cleaned', data=df, order=order)
plt.xticks(rotation=45)
plt.title('Histogram zmiennej Stay', fontsize=16)
plt.xlabel('Czas pobytu w szpitalu', fontsize=12)
plt.ylabel('Liczba przypadków', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
